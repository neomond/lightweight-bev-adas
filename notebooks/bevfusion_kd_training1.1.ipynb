```python
# CELL 5: Attempt REAL BEVFusion teacher via isolated CUDA 11.3 conda env
# Colab's driver can't be changed, but drivers are backward-compatible with
# older CUDA runtimes, so we install CUDA 11.3 toolkit inside conda instead.
import os, subprocess, time

CONDA_PREFIX = '/usr/local/miniconda'
ENV_NAME     = 'bevfusion'
ENV_PYTHON   = f'{CONDA_PREFIX}/envs/{ENV_NAME}/bin/python'
ENV_PIP      = f'{CONDA_PREFIX}/envs/{ENV_NAME}/bin/pip'

TEACHER_MODE = 'mock'  # default; flipped to 'real' only if every step below succeeds

def run(cmd, timeout=1800):
    print(f'$ {cmd}')
    r = subprocess.run(cmd, shell=True, capture_output=True, text=True, timeout=timeout)
    if r.returncode != 0:
        print(r.stdout[-2000:])
        print(r.stderr[-2000:])
    return r.returncode == 0

try:
    t0 = time.time()

    if not os.path.exists(CONDA_PREFIX):
        print('Installing Miniconda...')
        assert run('wget -q https://repo.anaconda.com/miniconda/Miniconda3-latest-Linux-x86_64.sh')
        assert run(f'bash Miniconda3-latest-Linux-x86_64.sh -b -f -p {CONDA_PREFIX}')

    if not os.path.exists(ENV_PYTHON):
        print('Creating bevfusion conda env (python 3.8)...')
        assert run(f'{CONDA_PREFIX}/bin/conda create -n {ENV_NAME} python=3.8 -y')

        print('Installing PyTorch 1.10.1 + CUDA 11.3 toolkit into env...')
        assert run(
            f'{CONDA_PREFIX}/bin/conda install -n {ENV_NAME} -c pytorch -c nvidia '
            f'pytorch==1.10.1 torchvision==0.11.2 cudatoolkit=11.3 -y',
            timeout=1200
        )

        print('Installing mmcv-full (cu113/torch1.10.0)...')
        assert run(
            f'{ENV_PIP} install mmcv-full==1.4.0 '
            f'-f https://download.openmmlab.com/mmcv/dist/cu113/torch1.10.0/index.html',
            timeout=1800
        )

        print('Installing mmdet / mmsegmentation...')
        assert run(f'{ENV_PIP} install mmdet==2.20.0 mmsegmentation==0.20.2')

    # Build BEVFusion's custom CUDA ops - this is the step most likely to fail
    BEVFUSION_DIR = '/content/bevfusion'
    if not os.path.exists(BEVFUSION_DIR):
        print('Cloning BEVFusion...')
        assert run(f'git clone https://github.com/mit-han-lab/bevfusion.git {BEVFUSION_DIR}')

    print('Building BEVFusion custom ops against CUDA 11.3 (can take 20-40 min)...')
    ok = run(f'cd {BEVFUSION_DIR} && {ENV_PYTHON} setup.py develop', timeout=3600)

    if ok:
        # Sanity check: can the env actually import the compiled package + see the GPU
        check = subprocess.run(
            f'{ENV_PYTHON} -c "import torch, mmdet3d; print(torch.cuda.is_available())"',
            shell=True, capture_output=True, text=True
        )
        if check.returncode == 0 and 'True' in check.stdout:
            TEACHER_MODE = 'real'
            print(f'\u2705 Real BEVFusion teacher ready. Setup took {(time.time()-t0)/60:.1f} min')
        else:
            print('mmdet3d imported but GPU check failed, falling back to mock')
    else:
        print('Custom CUDA op build failed, falling back to mock')

except Exception as e:
    print(f'Setup failed: {e}')

print(f'\nTeacher mode: {TEACHER_MODE}')
```

```python
# CELL 6: Generate teacher cache
import os
from pathlib import Path
os.chdir('/content/repo')

train_dir = Path(CACHE_PATH) / 'train'
val_dir   = Path(CACHE_PATH) / 'val'
train_count = len(list(train_dir.glob('*.pt'))) if train_dir.exists() else 0
val_count   = len(list(val_dir.glob('*.pt')))   if val_dir.exists()   else 0
print(f'Existing cache: {train_count} train, {val_count} val')

if train_count >= 324 and val_count >= 80:
    print('Cache complete. Skip to Cell 7.')
else:
    if TEACHER_MODE == 'real':
        python_bin = ENV_PYTHON      # conda env's python - has real mmdet3d
        mock_flag  = ''
        print('Using REAL BEVFusion teacher (conda env)')
    else:
        python_bin = 'python'        # normal Colab kernel python
        mock_flag  = '--mock'
        print('Using mock teacher')

    for split in ['train', 'val']:
        print(f'Caching {split}...')
        os.system(f'{python_bin} scripts/cache_teacher_outputs.py '
                  f'--config configs/student.yaml {mock_flag} '
                  f'--split {split} --cache-dir data/teacher_cache')

    train_count = len(list(train_dir.glob('*.pt')))
    val_count   = len(list(val_dir.glob('*.pt')))
    print(f'Cache complete: {train_count} train, {val_count} val')
```

```python
# CELL 9: Results summary table
import torch
from pathlib import Path

print('=' * 55)
print('  DISSERTATION RESULTS TABLE')
print('=' * 55)

checkpoints = {
    'Baseline (no KD), 20ep':  f'{CKPT_PATH}/baseline_20ep_best.pth',
    'Student + KD, 20ep':      f'{CKPT_PATH}/kd_real_best.pth',
}

results = {}
for name, path in checkpoints.items():
    if Path(path).exists():
        ckpt = torch.load(path, map_location='cpu', weights_only=False)
        results[name] = ckpt['val_loss']
        print(f'  {name:<30} val_loss={ckpt["val_loss"]:.4f}  epoch={ckpt["epoch"]}')
    else:
        print(f'  {name:<30} NOT YET TRAINED')

if len(results) == 2:
    k = list(results.keys())
    delta = results[k[0]] - results[k[1]]
    pct   = delta / results[k[0]] * 100
    print()
    print(f'  KD improvement: {delta:+.4f}  ({pct:.1f}% val loss reduction)')
    if delta > 0:
        print('  KD improved over baseline')
    else:
        print('  KD did not improve (expected with mock teacher)')
print('=' * 55)
```
